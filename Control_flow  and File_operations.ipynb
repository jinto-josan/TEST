### CONDITIONAL (if elif else)

- Condition can be formed using <,=,<=,>,>=,==,!=
- Conditions can be joined using and , or, not
- Conitions based on iterables using in , not in
- All integers other than 0 are considered true
- Any condition that gives bool output can be used


```python
if 3>2:
    print(True)
```

```python
if 1>2:
    print(True)
else:
    print(False)
```

```python
a=3
b=5
if a>b:
    print("a is greater than b")
elif a<b:
    print("b is greater than a")
else:
    print("a is equal to b")
```

```python
a=5
b=6
c=2
if a<=b and a<=c:
    print("Smallest is a")
elif b<=c and b<=a:
    print("Smallest is b")
else:
    print("Smallest is a")
```

```python
lst=[1,2,3,4,5]
if 2 in lst:
    print("2 is present")
```

```python
a=-1
if a:
    print(a)
```

```python
def check_odd(a):
    return a%2
if check_odd(3):
    print("odd")
```

### LOOPS
- For loop (using range, list comprehension, for else)
- while loop (doesnt iterate only takes condition)

```python
for i in range(1,5,2):
    print(i)
```

```python
lst = [1,2,3,4,5]
for i in lst:
    print(i,end=" ")
```

```python
[i for i in range(1,10) if i%2==0]
```

```python
lst=list()
for i in range(1,10):
    if(i%2==0):
        lst.append(i)
lst
```

```python
for i in range(10,15):
    print(i)
else:
    print("I am not in loop anymore")
```

```python
while i in range(1,10):
    print(i)
```

```python
start=0
end=10
i=0
while i>=start and i<end:
    print(i,end=" ")
    i+=1
```

### File Operation

- Various modes of file open
    - r  Opens a file for reading. (default)
    - w  Opens a file for writing. Creates a new file if it does not exist or Reomves file content if file exists.
    - x  Opens a file for exclusive creation. If the file already exists, the operation fails.
    - a  Opens a file for appending. Creates a new file if it does not exist.
    - '+' Opens a file for updating (reading and writing)
    - t Opens in text mode. (default)
    - b Opens in binary mode.
    
    
- Various methods
    - open
    - read
    - readline
    - readlines
    - write
    - writelines
    - tell
    - seek
    - flush
    - close
    
   

```python
file=open('new.txt','w') #it trunctuates all that was written earlier. Then its the same process of appending
```

```python
file.write("James\n") # Just writing wont update the file
```

```python
file.flush() #Flushing moves the lines from buffer to the file as write operations are costly
```

```python
for i in ['Alpha','Beta','Gamma']: #Writing multiple lines using for loop
    file.write(i+"\n")
```

```python
file.writelines(["Alpha1\n","Alpha2\n"]) #Yet another function to write multiple lines
```

```python
file.close() #Closing a file flushes everything in the file
```

```python
file=open('new.txt','a') #Apppending doesnot trunctuates the file but appends line at end
for i in range(1,10):
    file.write("{} , ".format(i))
file.close()
```

```python
file=open('new.txt','r') #Reading operation
file.read() #read without any value reads entire file
```

```python
file.tell() #gives the current location of read cursor
```

```python
file.seek(0) #to move to a particular location in a file
file.readline() #reading line
```

```python
file.read(2)#reading two character in second line as our current read cursor is at the second line
```

```python
file.readlines()#read lines from the read cursor position to end and returns as list
```

```python
file.close()
```

```python
file=open('new.txt','r+')#opens the file with trait of reading and writing
file.readline()
```

```python
file.write("\nThank you\n")
file.close()
```

```python
# with automatically closes the file and thus we dont have to explictly define that
with open('new.txt','r') as f:
    print(f.read())
```
